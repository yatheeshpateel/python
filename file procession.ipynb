```python
import pickle
class Employee:
    def __init__(self,name,age,sal,dept):
        self.ename = name
        self.eage = age
        self.esal = sal
        self.edept = dept
    def disp(self):
        print(self.ename)
        print(self.eage)
        print(self.esal)
        print(self.edept)

e = Employee("rama",25,1000000,"Data Engineering")
f = open("Employee.txt","wb")
pickle.dump(e,f)
f.close()
print('object is stored in file')
```

```python
f = open("Employee.txt","rb")
e = pickle.load(f)
e.disp()
print('object is retrive from file')
f.close()
```
