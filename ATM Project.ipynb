# Generate a Bill 

```python
menu = {
    "Idli": 20.00,
    "Dosa": 50.00,
    "Poori": 25.00,
    "fruit_salad": 60.00
}

ordered_breakfasts = ["Idli", "Dosa",  "fruit_salad"]

total_cost = 0.0

print("Hotel Bill")
print("----------")
print("Item          Cost ")

for breakfast in ordered_breakfasts:
    if breakfast in menu:
        cost = menu[breakfast]
        total_cost += cost
        item_name = breakfast + '          ' 
        item_name = item_name[:12]  
        print(item_name + " " + str(cost))

        
print("----------")
print("Total Cost:   " + str(total_cost))
```
